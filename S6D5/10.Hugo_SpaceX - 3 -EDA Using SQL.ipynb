## Tarea 2: Queries con SQL

### Ejercicio 1

##### Muestra las diferentes plataformas de lanzamiento:

```python
#Rellena la variable query con tu consulta
query = """ select LaunchSite from SPACEXTBL """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 2

##### Muestra 5 lanzamientos cuya plataforma empiece por 'CCS'


```python
#Rellena la variable query con tu consulta

query = """ select LaunchSite from SPACEXTBL
            where LaunchSite like 'CCS%'
            limit 5 """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 3

##### Mustra el la masa total transportada por los lanzamientos realizados donde el cliente es NASA (CRS)


```python

query = """ select sum(PayLoadMass), Customers from SPACEXTBL
            group by Customers
            having Customers like 'NASA'
            """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 4

##### Muestra la masa media tranportada por lanzamientos con serie que comienza por B1

```python
query = """ select avg(PayLoadMass) from SPACEXTBL
            where Serial like 'B1%'
            """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 5

##### Muestra la fecha donde se realizo el primer aterrizaje exitoso sobre terreno (TRUE RTLS)

*Pista:Utiliza la función min*


```python
query = """ select min(Date) as 'Día primer éxito aterrizaje en tierra', LandingOutcome, Class from SPACEXTBL
            where LandingOutcome='TRUE RTLS'              #and Class=1 no necesario
            """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 6

##### Muestra las series de los cohetes que aterrizaron sobre portaviones (TRUE ASDS) con una masa entre 4000 y 6000

```python
query = """ select Serial, LandingOutcome from SPACEXTBL
            where LandingOutcome='TRUE ASDS' and PayloadMass between 4000 and 6000
            """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 7

##### Muestra el numero de lanzamientos que completaron su mision correctamente y aquellos que fallaron

```python

#Verisón1 sumar todo y agrupando por la clase
query = """ select count(Serial) as 'Número de fracasos y éxitos' from SPACEXTBL
            group by Class
            """
df  = pd.read_sql(query, con = engine )
df

#Verisón2 suamndo la clase
query = """ select sum(Class) as Éxitos, sum(case when Class=0 then 1 end) as 'Fracasos' from SPACEXTBL
            """

df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 8

##### Muestra las series de los cohetes que transportaron la masa máxima

```python

#verion1 ordenanado por la carga maxima
query = """ select Serial, PayloadMass from SPACEXTBL
            order by PayloadMass desc
            """
df  = pd.read_sql(query, con = engine )
df


#verison2 calcular la carga maxima, pero solo muestra uno si están empatdos

query = """ select Serial, max(PayloadMass) from SPACEXTBL
            """
df  = pd.read_sql(query, con = engine )
df

#verison3 saber cual e sla carga maxima y hacer un where
query = """ select Serial, PayloadMass from SPACEXTBL
            where PayloadMass=15600
            """
df  = pd.read_sql(query, con = engine )
df

#version4 verison3 pero clacuadno el valor maximo en el where
query = """ select Serial, PayloadMass from SPACEXTBL
            where PayloadMass=(select max(PayloadMass) from SPACEXTBL)
            """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 9

##### Muestra los el resultados del aterrizaje, la serie del cohete y las plataforma de lanzamiento de aquellos lanzamientos fallidos o no intentados en el año 2015


```python
query = """ select LandingOutcome, Serial, LaunchSite, Year(Date) as Anyo from SPACEXTBL
            where Year(Date)='2015' and LandingOutcome like 'None%' or LandingOutcome like '%False%'
            """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 10

##### Muestra los resultados y la frecuencia de estos resultados entre 2010-06-04 y 2017-03-20 ordenados en orden descendiente según la frecuencia


```python
query = """ select count(LandingOutcome) as Frecuencia, LandingOutcome from SPACEXTBL
            where   (year(Date)='2010' and month(Date)='6' and day(Date)>'3')
                    or (year(Date)='2010' and month(Date)>'6')
                    or (year(Date)>'2010' and year(Date) <'2017')
                    or (year(Date)='2017' and month(Date)<'4')
                    or (year(Date)='2017' and month(Date)='3' and day(Date)<'20')
            group by LandingOutcome
            order by Frecuencia desc
            """
df  = pd.read_sql(query, con = engine )
df
```
